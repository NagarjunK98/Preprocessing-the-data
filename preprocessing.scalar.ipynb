```python
from sklearn import preprocessing
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#reading the csv file of california house dataset
# Considering 2 features out of 8 because it will be easy to understand scaling techique
# Two features considered are in different distibution
data=pd.read_csv("C:/Users/nagarjun/Desktop/cal_housing.csv",usecols=['medianIncome','households'])
```

```python
#To display first 10 rows
data.head(10)
```

```python
# To know the dimension of data considered
data.shape
```

```python
# Considering two different data Series to plot original data
X1=data['medianIncome']
X2=data['households']
```

```python
#plotting of original data
# As we can see features values are different scale
plt.xlabel("Median_income")
plt.ylabel("House_hold")
plt.plot(X1,X2,'.',markersize=5)
```

```python
#Implementing StandardScaler
# StandardScaler implements zero mean and unit standard deviation
scaler=preprocessing.StandardScaler()
X_scaled=scaler.fit_transform(data)
```

```python
# plotting Scaled values
plt.xlabel("Median_income")
plt.ylabel("House_hold")
plt.plot(X_scaled[:,1:],X_scaled[:,0:1],'.',markersize=5)
```

```python
#Implementing MinMaxScaler
# MinMax gives the values in range[0,1]
scaler=preprocessing.MinMaxScaler()
X_scaled=scaler.fit_transform(data)
```

```python
# plotting Scaled values
plt.xlabel("Median_income")
plt.ylabel("House_hold")
plt.plot(X_scaled[:,1:],X_scaled[:,0:1],'.',markersize=5)
```

```python
#Implementing MaxAbsScaler
# As data consists of positive values getting scaled range in[0,1] only same as MaxMinScaler.
#If data consists of negative then scaled value distribution will be different 
scaler=preprocessing.MaxAbsScaler()
X_scaled=scaler.fit_transform(data)
```

```python
# plotting Scaled values
plt.xlabel("Median_income")
plt.ylabel("House_hold")
plt.plot(X_scaled[:,1:],X_scaled[:,0:1],'.',markersize=5)
```

```python
#Implementing RobustScaler
# This scaling technique is based on percentile it will select values with in given range
scaler=preprocessing.RobustScaler(quantile_range=(10, 80))
X_scaled=scaler.fit_transform(data)
```

```python
# plotting Scaled values
plt.xlabel("Median_income")
plt.ylabel("House_hold")
plt.plot(X_scaled[:,1:],X_scaled[:,0:1],'.',markersize=5)
```

```python
#Implementing PowerTransformer(method='Yeo-johnson')
# If data consists of both positive and negative values then we can use this method
scaler=preprocessing.PowerTransformer(method='yeo-johnson')
X_scaled=scaler.fit_transform(data)
```

```python
# plotting Scaled values
plt.xlabel("Median_income")
plt.ylabel("House_hold")
plt.plot(X_scaled[:,1:],X_scaled[:,0:1],'.',markersize=5)
```

```python
#Implementing PowerTransformer(method='box-cox')
# If data consists of only positive values then we can use this method
scaler=preprocessing.PowerTransformer(method='box-cox')
X_scaled=scaler.fit_transform(data)
```

```python
# plotting Scaled values
plt.xlabel("House_hold")
plt.ylabel("Median_income")
plt.plot(X_scaled[:,1:],X_scaled[:,0:1],'.',markersize=5)
```

```python
#Implementing QuintileTransformer(output_distrubution='normal')
# this method tries to gives Gaussian distribution values
scaler=preprocessing.QuantileTransformer(output_distribution='normal')
X_scaled=scaler.fit_transform(data)
```

```python
# plotting Scaled values
plt.xlabel("House_hold")
plt.ylabel("Median_income")
plt.plot(X_scaled[:,1:],X_scaled[:,0:1],'.',markersize=5)
```

```python
#Implementing QuintileTransformer(output_distrubution='uniform')
# this method gives uniform values
scaler=preprocessing.QuantileTransformer(output_distribution='uniform')
X_scaled=scaler.fit_transform(data)
```

```python
# plotting Scaled values
plt.xlabel("House_hold")
plt.ylabel("Median_income")
plt.plot(X_scaled[:,1:],X_scaled[:,0:1],'.',markersize=5)
```

```python
#Implementing Normalizer
scaler=preprocessing.Normalizer()
X_scaled=scaler.fit_transform(data)
```

```python
# plotting Scaled values
plt.xlabel("House_hold")
plt.ylabel("Median_income")
plt.plot(X_scaled[:,1:],X_scaled[:,0:1],'.',markersize=5)
```
